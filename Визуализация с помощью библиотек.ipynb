Распределение должно быть по всем образцам, без разделения на вид.

```python
df5= pd.read_csv('C:/Users/nikit/Downloads/iris.csv', index_col=0)
```

```python
df5.head()
```

```python
# xтобы построить на 1-ом графике распределения для каждого из параметров, можно воспользоваться петлёй(циклом):

for column in df5.columns[:-1]:  # Пропускаем последний столбец (вид):
    sns.displot(df5, x=column, kde=True)
    sns.kdeplot(data=df5)
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.show()
```

```python
sns.pairplot(df5,  vars=df5.columns[:-1], diag_kind='kde')
```

```python
for column in df5:
    sns.kdeplot(df5[str(column)])
    sns.kdeplot(data=df5)
```

Итог:
- Sepal width и sepal length имеют унимодальное распределение;
- Petal width и petal length имеют бимодальное распределение;
- Petal length имеет наибольший размах значению.

Рассмотрим длину лепестков (petal length) подробнее и воспользуемся для этого violin плотом.
