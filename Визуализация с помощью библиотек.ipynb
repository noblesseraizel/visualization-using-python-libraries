```python
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
students_perfomance = pd.read_csv('C:/Users/nikit/Downloads/StudentsPerformance.csv')
```

```python
students_perfomance.head(3)
```

```python
students_perfomance = students_perfomance.rename(columns={'parental level of education': 'parental_level_of_education', 'test preparation course': 'test_preparation_course', 'math score': 'math_score', 'reading score': 'reading_score', 'writing score': 'writing_score'})
# переименуем колонки, для более быстрого их выделения
```

```python
students_perfomance.math_score.hist() 
# быстрый способ оценить распределение в данных
```

```python
students_perfomance.plot.scatter(x='math_score', y='reading_score') 
# создаем точечный график двух переменных
```

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_perfomance, fit_reg=False)
# hue группирующая переменная, fit_reg регрессионная прямая
ax.set_xlabels('math score')
ax.set_ylabels('reading score')
# тот же график с разбивкой по группе 'gender'
```
